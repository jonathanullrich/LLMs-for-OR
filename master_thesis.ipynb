## Utils

````python
class Result(Enum):
    ACCEPT = 0
    WRONG_ANSWER = 1
    RUNTIME_ERROR = 2
    COMPILE_ERROR = 3
````

````python
def strip_code(text):
  """Removes non-code blocks from generated text and concatenates it"""
  pattern = r'```python\s*([\s\S]*?)\s*```'
  matches = re.findall(pattern, text)
  concatenated_code = '\n'.join(matches)
  return concatenated_code
````

````python
def check_imports(code):
    """Checks imports in generated code and installs them, if necessary"""
    try:
      tree = ast.parse(code)
    except Exception as e:
      return Result.COMPILE_ERROR
    imps = set()
    for node in ast.walk(tree):
        if isinstance(node, ast.Import):
            for alias in node.names:
                imps.add(alias.name)
        elif isinstance(node, ast.ImportFrom):
            module_name = node.module
            for alias in node.names:
                imps.add(f"{module_name}.{alias.name}")
    for imp in imps:
        imp = imp.split(".")[0]
        try:
            importlib.import_module(imp)
        except ImportError:
            #print(f"{imp} cannot be imported - attempting to install with pip.")
            try:
                subprocess.check_call(['pip', 'install', imp])
            except Exception as e:
                return Result.COMPILE_ERROR
````

````python
def run_code(generated_code):
  """Compiles and executes generated code. Returns either solution number or <Result>"""
  vars = { "solution": None }
  if generated_code == "":
    return Result.COMPILE_ERROR
  try:
      warnings.filterwarnings("ignore")
      compiled_code = compile(generated_code, '<string>', 'exec')
      exec(compiled_code, vars)
      warnings.filterwarnings("default")
  except (SyntaxError, IndentationError, NameError, TypeError, ValueError) as e:
      print(f"Compile error: {e}")
      return Result.COMPILE_ERROR
  except Exception as e:
      print(f"Runtime error: {e}")
      return Result.RUNTIME_ERROR
  return vars["solution"] if isinstance(vars["solution"], (int, float)) else Result.RUNTIME_ERROR
````

````python
def check_solution(pred_sol, ground_truth):
  # pred_sol = round(pred_sol, 3)
  return Result.ACCEPT if pred_sol == ground_truth else Result.WRONG_ANSWER
````

## Standard Prompting

````python
def standard_prompting(problem, shots=0): # shots in arg is ignored - zero shot
  """Prompts LLM, sanitizes and logs output. Returns a <Result>"""

  system_prompt = "You are an operations research expert. Write python code to solve the given problem. Assign the final output to a variable 'solution'. Only assign one number."
  problem_description = f"Consider the following problem: {problem['description']}"

  response = client.chat.completions.create(
    model="gpt-4",
    messages=[
      {"role": "system", "content": system_prompt},
      {"role": "user", "content": problem_description},
    ]
  )

  generated_code = strip_code(response.choices[0].message.content)
  import_error = check_imports(generated_code)
  if import_error is not None:
    res = import_error
  else:
    res = run_code(generated_code)

  data = {
    "System prompt": system_prompt,
    "Problem description": problem_description,
    "Output": response.choices[0].message.content,
    "Code": generated_code,
    "Solution": res if isinstance(res, (int, float)) else res.name,
  }

  uid = uuid.uuid4()
  uid_str = str(uid)

  file_path = f"{save_dir}standard/{problem['name']}-{uid_str}.json"
  with open(file_path, 'w') as json_file:
    json.dump(data, json_file, indent=4)

  return res
````

````python
def cot_prompting(problem, shots=0):
  """Prompts LLM with CoT, sanitizes and logs output. Returns a <Result>"""

  system_prompt = "You are an operations research expert. Write python code to solve the given problem. Assign the final output to a variable 'solution'. Only assign one number."
  problem_description = f"Consider the following problem: {problem['description']}"

  if shots == 1:
    for example in examples: # pick matching example for problem class
      if example["class"] == "linear":
        few_shot_examples = "Here is an example how to solve a similar operations research problem:\n" + '\n'.join(["Problem: " + example["description"], example["steps"]])
  elif shots == 3:
    few_shot_examples = "Here are a few examples how to solve operations research problems:\n" + '\n---\n'.join(['\n'.join(["Problem: " + example["description"], example["steps"]]) for example in examples]) # concatenates examples in single string
  else: # 0-shot
    few_shot_examples = "Let's think step by step."

  response = client.chat.completions.create(
    model="gpt-4",
    messages=[
      {"role": "system", "content": system_prompt},
      {"role": "user", "content": problem_description},
      {"role": "system", "content": few_shot_examples},
    ]
  )
  generated_code = strip_code(response.choices[0].message.content)
  import_error = check_imports(generated_code)
  if import_error is not None:
    res = import_error
  else:
    res = run_code(generated_code)

  data = {
    "System prompt": system_prompt,
    "Problem description": problem_description,
    "Few-Shot examples": few_shot_examples,
    "Output": response.choices[0].message.content,
    "Code": generated_code,
    "Solution": res if isinstance(res, (int, float)) else res.name,
  }
  uid = uuid.uuid4()
  uid_str = str(uid)

  file_path = f"{save_dir}cot/{shots}-shots-{problem['name']}-{uid_str}.json"
  with open(file_path, 'w') as json_file:
    json.dump(data, json_file, indent=4)

  return res
````

## Chain of Thought with Self-Consistency

````python
def cotsc_prompting(problem, shots=0):
  """Prompts LLM with CoT-SC, sanitizes and logs output. Returns a <Result>"""

  system_prompt = "You are an operations research expert. Write python code to solve the given problem. Assign the final output to a variable 'solution'. Only assign one number."
  problem_description = f"Consider the following problem: {problem['description']}"

  if shots == 1:
    for example in examples:
      if example["class"] == problem["class"]: # pick matching example for problem class
        few_shot_examples = "Here is an example how to solve a similar operations research problem:\n" + '\n'.join(["Problem: " + example["description"], example["steps"]])
  elif shots == 3:
    few_shot_examples = "Here are a few examples how to solve operations research problems:\n" + '\n---\n'.join(['\n'.join(["Problem: " + example["description"], example["steps"]]) for example in examples]) # concatenates examples in single string
  else: # 0-shot
    few_shot_examples = "Let's think step by step."

  response = client.chat.completions.create(
    model="gpt-4",
    messages=[
      {"role": "system", "content": system_prompt},
      {"role": "user", "content": problem_description},
      {"role": "system", "content": few_shot_examples},
    ],
    temperature=0.7,
    n=5
  )

  sol_ls = []
  res_ls = [choice.message.content for choice in response.choices]
  code_ls = []
  for sample in res_ls:
    generated_code = strip_code(sample)
    code_ls.append(generated_code)
    import_error = check_imports(generated_code)
    if import_error is not None:
      return import_error # fix
    tmp = run_code(generated_code)
    if isinstance(tmp, (int, float)):
      sol_ls.append(tmp)

  value_counts = {value: sol_ls.count(value) for value in set(sol_ls)}
  max_count = max(value_counts.values())
  max_value_candidates = [value for value, count in value_counts.items() if count == max_count]
  sol = random.choice(max_value_candidates) # Randomly choose if tie in majority

  data = {
    "System prompt": system_prompt,
    "Problem description": problem_description,
    "Few-Shot examples": few_shot_examples,
    "Output": res_ls,
    "Code": code_ls,
    "Solutions": sol_ls,
    "Chosen Solution": sol,
  }
  uid = uuid.uuid4()
  uid_str = str(uid)

  file_path = f"{save_dir}cotsc/{shots}-shots-{problem['name']}-{uid_str}.json"
  with open(file_path, 'w') as json_file:
    json.dump(data, json_file, indent=4)

  return sol
````

## Tree of Thought

````python
def tot_prompting(problem, shots=0):
  """Prompts LLM with ToT, sanitizes and logs output. Returns a <Result>"""

  system_prompt = "You are an operations research expert. Keep your answers concise."
  problem_description = f"Consider the following problem: {problem['description']}"

  if shots == 1:
    for example in examples:
      if example["class"] == problem["class"]: # pick matching example for problem class
        few_shot_examples = "Here is an example how to solve a similar operations research problem:\n" + '\n'.join(["Problem: " + example["description"], example["steps"]])
  elif shots == 3:
    few_shot_examples = "Here are a few examples how to solve operations research problems:\n" + '\n---\n'.join(['\n'.join(["Problem: " + example["description"], example["steps"]]) for example in examples]) # concatenates examples in single string
  else: # 0-shot
    few_shot_examples = "Let's think step by step."

  cot_prompt_plan = "Determine the best approach to solve the problem. Do not write code yet."
  cot_prompt_init = "Write python code to model the problem. Do not solve the problem yet, only make sure to model the problem correctly. Use libraries, if needed."
  cot_prompt_solve = "Consider the python code to model the problem. Extend the code to solve the problem. Assign the final output to a variable 'solution'. Only assign one number."

  tot_breadth = 3
  cot_prompts = [cot_prompt_plan, cot_prompt_init, cot_prompt_solve] # tot depth = 3

  sequential_logs = []
  previous_layer_output = "" # succesively build context based on previous layers' output

  for layer_prompt in cot_prompts:

    # generate thought by sampling from a CoT prompt
    response = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": problem_description},
        {"role": "system", "content": few_shot_examples},
        {"role": "system", "content": previous_layer_output},
        {"role": "system", "content": layer_prompt},
      ],
      temperature=0.7,
      n=tot_breadth
    )

    layer_samples = [f'Choice {i}: ' + choice.message.content for i, choice in enumerate(response.choices)]
    sequential_logs.append("\n".join(layer_samples)) # log samples

    # evaluate thought by voting
    vote_prompt = "You are given the following choices for the current step. Decide which choice is most promising. Analyze each choice in detail, then conclude in the last line 'The best choice is {i}', where i the integer id of the choice."

    selection = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": problem_description},
        {"role": "system", "content": vote_prompt},
        {"role": "user", "content": previous_layer_output},
        {"role": "user", "content": "\n".join(layer_samples)},
      ],
      temperature=0,
      n=1
    )

    sequential_logs.append(selection.choices[0].message.content) # log selection

    pattern = r'The best choice is (\d+)'
    match = re.search(pattern, selection.choices[0].message.content)
    if match:
        number_str = match.group(1)
        try:
          choice = int(number_str)
          previous_layer_output += layer_samples[choice] # might not have 3 previous output here then
          sequential_logs.append("Choice: " + str(choice)) # log choice
        except ValueError as e:
          return None


  generated_code = strip_code(previous_layer_output)
  import_error = check_imports(generated_code)
  if import_error is not None:
    res = import_error
  else:
    res = run_code(generated_code)

  data = {
    "System prompt": system_prompt,
    "Problem description": problem_description,
    "Output": sequential_logs,
    "Code": generated_code,
    "Solution": res if isinstance(res, (int, float)) else res.name,
  }
  uid = uuid.uuid4()
  uid_str = str(uid)

  file_path = f"{save_dir}tot/{shots}-shots-{problem['name']}-{uid_str}.json"
  with open(file_path, 'w') as json_file:
    json.dump(data, json_file, indent=4)

  return res
````
